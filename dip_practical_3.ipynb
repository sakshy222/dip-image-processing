```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load images in grayscale
img1 = cv2.imread('/content/img1.jpg', 0)
img2 = cv2.imread('/content/img2.jpg', 0)

# Resize if sizes are different
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# a) ADDITION of two images
add = cv2.add(img1, img2)

# b) SUBTRACTION of two images
sub = cv2.subtract(img1, img2)

# c) MEAN value of image
mean_value = np.mean(img1)
print("Mean intensity of Image 1:", mean_value)

# d) CHANGE BRIGHTNESS by changing mean value
# Increase brightness
bright = img1 + 50
bright = np.clip(bright, 0, 255)

# Decrease brightness
dark = img1 - 50
dark = np.clip(dark, 0, 255)

# Show results
print("Original Image")
cv2_imshow(img1)

print("Added Image")
cv2_imshow(add)

print("Subtracted Image")
cv2_imshow(sub)

print("Bright Image")
cv2_imshow(bright.astype(np.uint8))

print("Dark Image")
cv2_imshow(dark.astype(np.uint8))
```
